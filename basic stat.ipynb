# Data Visualization

```python
for column in numerical_cols:
    plt.figure()
    numerical_cols[column].hist(bins=30)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import seaborn as sns

for column in numerical_cols:
    plt.figure()
    sns.boxplot(y=numerical_cols[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

# Bar Chart Analysis for Categorical Column

```python
# Identify chategorical columns
df.select_dtypes(include=['object']).columns
```

```python
cat_df= df[df.select_dtypes(include=['object']).columns]
cat_df.drop(['Date'],axis=1,inplace=True)
cat_df
```

```python
# Get the list of categorical columns
categorical_columns = cat_df.select_dtypes(include=['object']).columns

# Create a single figure with multiple subplots
fig, axes = plt.subplots(len(categorical_columns), 1, figsize=(12, len(categorical_columns)*6))

# Plot countplots for each categorical column
for i, column in enumerate(categorical_columns):
    sns.countplot(data=cat_df, x=column, ax=axes[i])
    axes[i].set_title(f'Frequency of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# Standardization of Numerical variables

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
standardized_data = scaler.fit_transform(numerical_cols)
column_names = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Creating DataFrame with column names
standardized_data = pd.DataFrame(scaler.fit_transform(numerical_cols), columns=column_names)
standardized_data
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12,12 ))

sns.histplot(standardized_data['Volume'], kde=True, ax=axes.ravel()[0], color='skyblue')
sns.histplot(standardized_data['Avg Price'], kde=True, ax=axes.ravel()[1], color='skyblue')
sns.histplot(standardized_data['Total Sales Value'], kde=True, ax=axes.ravel()[2], color='skyblue')
sns.histplot(standardized_data['Discount Rate (%)'], kde=True, ax=axes.ravel()[3], color='skyblue')
sns.histplot(standardized_data['Discount Amount'], kde=True, ax=axes.ravel()[4], color='skyblue')
sns.histplot(standardized_data['Net Sales Value'], kde=True, ax=axes.ravel()[5], color='skyblue')

plt.tight_layout()
plt.show()
```
